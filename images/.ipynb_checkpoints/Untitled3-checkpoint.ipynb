```python
def read(filename="data.txt"):
    scores = {}
    with open(filename) as f:
        for line in f:
            s,i = line.split(",")
            scores[s.strip()] = int(i.strip())
    return scores
```

```python
import operator
scores = read()
```

```python


def test(val):
    if val in scores:
        return scores[val]
    part0 = "https://runner.team-lab.com/q?str="
    part1 = "&token=npfEF2T59ITaHFAU7QhGBZbTGSlXrVEG"
    score = urllib.request.urlopen(part0 + val + part1).read()
    scores[val] = score
    time.sleep(1)
    return int(score)

def beam_search(n, l, h):
    
    best = {}
    
    for last in l:
        w_a = last + "A"
        s_a = test(w_a) 
        w_b = last + "B"
        s_b = test(w_b) 
        w_c = last + "C"
        s_c = test(w_c) 
        w_d = last + "D"
        s_d = test(w_d) 
        w_e = last + "E"
        s_e = test(w_e) 
        best[w_a] = int(s_a)
        best[w_b] = int(s_b)
        best[w_c] = int(s_c)
        best[w_d] = int(s_d)
        best[w_e] = int(s_e)
        
   
    n_l = [ k for k, v in sorted(best.items(), key=operator.itemgetter(1), reverse=True)]
    print(n,n_l)
    if(n >= 50):
        return 
    else:
        beam_search(n+1, n_l[:h], h)
        
    
        
```

```python
import sys


def query(a, b):
    print("? {} {}".format(a, b), flush=True)
    sys.stdout.flush()
    ans = input()
    return 1 if ans == ">" else 0


def addToThree(a, k):
    if query(a, k[1]):
        if query(a, k[0]):
            return [a] + k
        else:
            return [k[0], a, k[1], k[2]]
    else:
        if query(a, k[2]):
            return [k[0], k[1], a, k[2]]
        return [k[0], k[1], k[2], a]


def five(n):
    if n == 1: 
        return 'A'
    if n == 2: 
        'BA' if query('A', 'B') else 'AB'
    if n == 3: 
        big = 'A' if query('A', 'B') else 'B'
        small = 'B' if big == 'A' else 'A'
        
        if query('C', big):
            return small+big+'C'
        else:
            if query('C', small):
                return small+'C'+big
            return 'C'+small+big
    if n == 4:
        big1 = 'A' if query('A', 'B') else 'B'
        small1 = 'B' if big == 'A' else 'A'
        
        big2 = 'C' if query('C', 'D') else 'D'
        small2 = 'D' if big == 'C' else 'C'

        if query(big1, big2):
            return "".join(addToThree(small1, [big1, big2, small2]))
        return "".join(addToThree(small2, [big2, big1, small1]))
    else:
        # A > B
        big1 = "A" if query('A', 'B') else "B"
        small1 = "A" if big1 == "B" else "B"
        # C > D
        big2 = "C" if query('C', 'D') else "D"
        small2 = "C" if big2 == "D" else "D"
        # A > C
        big3 = big1 if query(big1, big2) else big2

        if big3 == big2:
            big1, big2 = big2, big1
            small1, small2 = small2, small1
            
        # A > C > D
        l = addToThree('E', [big1, big2, small2])
        idx = l.index(big1)
        if idx == 1:
            if query(small1, big2):
                return small2 + big2 + small1 + big1 + 'E' 
            elif query(small1, small2):
                return small2 + small1 + big2 + big1 + 'E'
            return small1 + small2 + big2 + big 1 + 'E'
        else:
            return "".join(addToThree(small1, l[1:])[::-1]) + big1


def merge(A, B):
    res = []
    while A or B:
        if A and B:
            if query(A[0], B[0]):
                res.append(B.pop(0))
            else:
                res.append(A.pop(0))
        elif A:
            res.extend(A)
            return res
        elif B: 
            res.extend(B)
            return res
    return res
        
    
def mergeSort(A):
    if len(A) <= 1:
        return A
    p = len(A)//2
    left = mergeSort(A[:p])
    right = mergeSort(A[p:])
    return merge(left, right)


N, Q = map(int, input().split())
if N != 5:
    s = "".join(mergeSort([chr(ord('A') + i) for i in range(N)]))
else:
    s = five(N)
print("! {}".format(s), flush=True)
sys.stdout.flush()
```

```python

def addToThree(a, k):
    if query(a, k[1]):
        if query(a, k[0]):
            return [a] + k
        else:
            return [k[0], a, k[1], k[2]]
    else:
        if query(a, k[2]):
            return [k[0], k[1], a, k[2]]
        return [k[0], k[1], k[2], a]


def five():
    # A > B
    big1 = "A" if query('A', 'B') else "B"
    # B < A
    small1 = "A" if big1 == "B" else "B"
    # C > D
    big2 = "C" if query('C', 'D') else "D"
    # D < C
    small2 = "C" if big2 == "D" else "D"
    # A > C
    big3 = big1 if query(big1, big2) else big2

    # A
    if big3 == big1:
        # A > C > D
        l = addToThree('E', [big3, big2, small2])
        idx = l.index(big3)
        if idx == 1:
            if query(small1, big2):
                return 'E' + big3 + small1 + big2 + small2
            elif query(small1, small2):
                return 'E' + big3 + big2 + small1 + small2
            return 'E' + big3 + big2 + small2 + small1
        else:
            return big3 + "".join(addToThree(small1, l[1:]))
    # C
    else:
        # C > A > B
        l = addToThree('E', [big3, big1, small1])
        idx = l.index(big3)
        if idx == 1:
            if query(small2, big1):
                return 'E' + big3 + small2 + big1 + small1
            elif query(small2, small1):
                return 'E' + big3 + big2 + small2 + small1
            return 'E' + big3 + big2 + small1 + small2
        else:
            return big3 + "".join(addToThree(small2, l[1:]))
```
